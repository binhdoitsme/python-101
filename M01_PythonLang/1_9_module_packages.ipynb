Commonly used modules & packages in Python:

```python
import os # OS function calls

print(os.listdir()) # list files in current directory
print(os.getcwd()) # get current directory
```

```python
import math

print(math.log(10)) # ln(10)
print(math.log(10, 10)) # log_10(10)

print(math.ceil(10.7))
print(math.floor(10.7))
print(round(10.7))
print(round(10.3))
```

```python
import requests

response = requests.get("https://google.com")

# explore public methods/properties
dir(response)
print(response.content)
print(response.text)
print(response.status_code)
print(response.headers)
print(response.url)
# print(response.json())
```

```python
import textwrap

s = """
    Hello World,
    This is my first text piece.
"""
print(s)
print("--------")
print(textwrap.dedent(s))
```
